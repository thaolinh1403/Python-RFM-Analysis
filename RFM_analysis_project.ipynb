```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import datetime as dt
%matplotlib inline
```

```python
data = pd.read_excel("ecommerce retail.xlsx")
```

```python
data=data[data['CustomerID'].notnull()]
print(data)
```

```python
data1 = data[['Country','CustomerID']].drop_duplicates()
data1.groupby('Country').size().sort_values(ascending=False)
```

```python
data = data[data['Country']=='United Kingdom']
data
```

```python
data = data[data['Quantity']>0]
data
```

```python
data = data[['CustomerID', 'InvoiceNo', 'InvoiceDate', 'Quantity', 'UnitPrice']]
data
```

```python
data = pd.DataFrame(data)
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
target_date = pd.to_datetime('2011-12-31')
data['Recency_Value'] = (target_date - data['InvoiceDate']).dt.days
data
```

```python
data['Revenue']=data['Quantity']*data['UnitPrice']
data
```

```python
rfm = data.groupby('CustomerID').agg(Recency = ('Recency_Value', 'min'), Frequency = ('InvoiceNo', 'nunique'), Monetary = ('Revenue', 'sum'))
rfm
```

```python
rfm.describe()
```

```python
bins_R = [20, 30, 50, 100, 200, 395]
bins_F = [0, 2, 5, 10, 50, 212]
bins_M = [-1, 300, 600, 1000, 2200, 260000]
labels_R = [5, 4, 3, 2, 1]
labels_F = [1, 2, 3, 4, 5]
labels_M = [1, 2, 3, 4, 5]

rfm['R_score'] = pd.cut(rfm['Recency'], bins=bins_R, labels=labels_R).astype(str)
rfm['F_score'] = pd.cut(rfm['Frequency'], bins=bins_F, labels=labels_F).astype(str)
rfm['M_score'] = pd.cut(rfm['Monetary'], bins=bins_M, labels=labels_M).astype(str)
rfm
```

```python
rfm['RFM Score']=rfm['R_score']+rfm['F_score']+rfm['M_score']
rfm['RFM Score']=rfm['RFM Score'].astype(int)
rfm=rfm.reset_index()
rfm_final = pd.DataFrame(rfm)
rfm_final
```

```python
seg = pd.read_excel("ecommerce retail.xlsx", sheet_name="Segmentation")
seg
```

```python
seg['RFM Score']=seg['RFM Score'].str.replace(' ', '')
seg['RFM Score']=seg['RFM Score'].str.replace(',', ', ')
seg
```

```python
new_data = []
for _, row in seg.iterrows():
    segment = row['Segment']
    rfm_scores = row['RFM Score'].split(', ')
    for rfm in rfm_scores:
        new_data.append({'RFM Score': rfm, 'Segment': segment})

result_df = pd.DataFrame(new_data)
result_df['RFM Score']=result_df['RFM Score'].astype(int)

result_df
```

```python
data = rfm_final.merge(result_df, how = "left", on='RFM Score')
data['R_score']=data['R_score'].astype(int)
data['F_score']=data['F_score'].astype(int)
data['M_score']=data['M_score'].astype(int)
data
```

```python
conditions = [
    data['Segment'].isin(['Champions', 'Loyal', 'Potential Loyalist']),
    data['Segment'].isin(['New Customers', 'Promising']),
    data['Segment'].isin(['Need Attention', 'About To Sleep', 'At Risk', 'Cannot Lose Them', 'Hibernating customers', 'Lost customers'])
]

values = ['Loyal customers', 'New customers', 'Old customers are about to leave']

data['Customer loyalty'] = np.select(conditions, values, default=None)
data
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['Monetary'], data['Recency'], alpha=0.5, color='tab:blue') 
plt.title('Monetary vs Recency')
plt.xlabel('Monetary')
plt.ylabel('Recency')
plt.grid(True)

plt.show()
```

#### The chart shows the correlation between Recent and Currency. Recent customers bring in more revenue than long-time customers. The number of new customers is quite large. There are still customers who spend a lot of money but have not returned for a long time, so there are programs specifically for this group of customers because they are still potential customers.

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['Monetary'], data['Frequency'], alpha=0.5, color='tab:blue')
plt.title('Frequency vs Monetary')
plt.xlabel('Monetary')
plt.ylabel('Frequenc')
plt.grid(True)

plt.show()
```

#### The chart shows the correlation between Frequency and Currency. The higher the customer's purchasing frequency, the greater the revenue.

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['Frequency'], data['Recency'], alpha=0.5, color='tab:blue') 
plt.title('Recency vs Frequency')
plt.xlabel('Frequency')
plt.ylabel('Recency')
plt.grid(True)

plt.show()
```

#### The chart shows the correlation between Recency and Frequency. Customers who have purchased recently are more likely to return to purchase than customers who have been purchasing for a long time.

```python
average_recency_by_segment = data.groupby('Segment')['Recency'].mean().reset_index()
average_recency_by_segment_sorted = average_recency_by_segment.sort_values('Recency')

custom_color = sns.color_palette("Blues")[3]
plt.figure(figsize=(8, 4))
ax = sns.barplot(x='Segment', y='Recency', data=average_recency_by_segment_sorted, color=custom_color)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8) 
plt.xticks(rotation=45)  
plt.title('Recency by Segment')
plt.xlabel('')
plt.ylabel('')

plt.show()
```

```python
average_recency_by_segment = data.groupby('Segment')['Frequency'].mean().reset_index()
average_recency_by_segment_sorted = average_recency_by_segment.sort_values('Frequency', ascending=False)

custom_color = sns.color_palette("Blues")[3]
plt.figure(figsize=(8, 4))
ax = sns.barplot(x='Segment', y='Frequency', data=average_recency_by_segment_sorted, color=custom_color)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8) 
plt.xticks(rotation=45)  
plt.title('Frequency by Segment')
plt.xlabel('')
plt.ylabel('')

plt.show()
```

#### The Champions segment is the most recently and most frequently purchased segment. This confirms that focusing on taking care of the Champions customer group is always the priority goal of the business.

```python
average_monetary_by_segment = data.groupby('Segment')['Monetary'].sum()
average_monetary_by_segment_sorted = average_monetary_by_segment.sort_values(ascending=False)

plt.figure(figsize=(5, 5))

wedges, texts, autotexts = plt.pie(
    average_monetary_by_segment_sorted,
    labels=average_monetary_by_segment_sorted.index,
    autopct='%1.1f%%',
    startangle=0, 
    textprops={'fontsize': 6}  
)

for text in texts:
    text.set_fontsize(6)  

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal') 
plt.title('Monetary by Segment')

plt.show()
```

#### From the chart, it can be seen that the segment that brings in the most revenue is Champions with 44.1%, followed by the Promises, Need Warning and Loyal segments.

```python
total_customers = len(data)

segment_counts = data['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Count']
segment_counts['Percentage'] = (segment_counts['Count'] / total_customers) * 100

colors = sns.color_palette("Blues_r", n_colors=len(segment_counts))

plt.figure(figsize=(7, 4))
ax = sns.barplot(x='Count', y='Segment', data=segment_counts, palette=colors)

for index, row in segment_counts.iterrows():
    ax.text(row["Count"], index, f'{row["Percentage"]:.1f}%', va='center', ha='left', fontsize=10, color='black')

plt.title("No. Customer by Segment")
ax.set_xlabel("")
ax.set_ylabel("")

plt.show()
```

#### Meanwhile, the number of Champions segment customers only accounts for 6.8% of the company's total customer base.

```python
customer_count = data['Customer loyalty'].value_counts()
monetary_sum = data.groupby('Customer loyalty')['Monetary'].sum()

fig, ax1 = plt.subplots(figsize=(6, 4))

ax1.set_xlabel('Customer loyalty')
ax1.set_ylabel('Customer Count', color='tab:blue')
ax1.bar(customer_count.index, customer_count, color='tab:blue', width=0.4)
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Monetary', color='tab:orange')
ax2.plot(monetary_sum.index, monetary_sum, color='tab:orange', marker='o')
ax2.tick_params(axis='y', labelcolor='tab:orange')

ax1.set_xlabel("")

plt.title('No Customers and Total Monetary by Customer Loyalty')

plt.show()
```

#### It can be seen that the number of customers belonging to the loyal customer group only accounts for about 19% of the total customer base but brings in many times more revenue than the other two groups.
> Therefore, it is necessary to identify the important customer group of the business, because that customer group will bear the majority of the business's revenue.
Here, the customer groups that SuperStore company needs to focus on are the loyal customer segment, the Promise segment, which belongs to the New customer group, and the Need Attention segment, which belongs to the Old customers are about to leave.

Based on the analysis above, both Recency and Frequency are related to revenue:
- Recent customers bring in more revenue than long-term customers
- The higher the customer's purchasing frequency, the greater the revenue

The following chart will more clearly analyze the relationship between the 3 indicators:

```python
loyalty_grouped = data.groupby('Customer loyalty')

loyalty_avg_scores = loyalty_grouped[['R_score', 'F_score', 'M_score']].mean()

colors = ['#6baed6', '#3182bd', '#08519c']
loyalty_avg_scores.plot(kind='bar', figsize=(8, 4), color=colors)

plt.xlabel("")
plt.xticks(rotation=0)
plt.title("R_score, F_score and M_score by Customer loyalty")
plt.ylabel("")
plt.legend(["R_score", "F_score", "M_score"])
           
plt.show()
```

#### For all 3 customer groups, F_score is always proportional to M_score, this proves that with high frequency of customer purchases, revenue will definitely be high. 
Especially in the retail industry, creating customer loyalty is extremely important. Focusing on purchase frequency helps companies maintain regular interactions with customers and create more frequent shopping habits, while increasing the likelihood of becoming loyal customers. 
So in this case SuperStore company should focus on the Frequency index, this will help the company maintain and increase value from current customers, while creating long-term growth opportunities.
